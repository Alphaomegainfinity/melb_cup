```python
#set-up driver
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#insert the website link that it is going to scrap
#url = 'https://racingaustralia.horse/FreeFields/Results.aspx?Key=2022Nov01%2CVIC%2CFlemington'
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#collect list of articles and preview list
section = soup.find('div', class_='page_content')

horse_lists = section.find_all('div', class_ ='state-specific-banner')
```

```python
# connect to SQL database 

protocol = 'postgresql'
username = config.username
password = config.password
host = 'localhost'
port = 5432
database_name = config.database_name
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
insp = inspect(engine)
```
